**1. Executive Summary**

Title: Ireland $CO_2$ Emissions Forecasting

Dataset: $CO_2$ Emissions in Ireland (1924-2023)

Source: https://ourworldindata.org/co2-emissions

Final Model: ARIMA (1, 0, 1)

Purpose:
This project investigates annual CO₂ emissions in Ireland from 1924 to 2023. The objective is to model and forecast CO₂ emissions using nonseasonal time series techniques. The final models considered include ARIMA(1,0,1) and an additive Exponential Smoothing (ETS) model. The ETS model was used because of its use by several sources in predicting future emissions. Forecasts suggest a gradual decline in emissions, consistent with observed post-2007 trends. These results support the view that recent environmental regulations and shifts in energy sources are having a sustained impact. The goal in creating forecasts such as this is to allow countries to track the impact of environmental regulations on the world and determine what steps are successful and what needs to be further changed.
